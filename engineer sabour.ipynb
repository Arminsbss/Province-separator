```python
from gettext import install
import tkinter as tk
from tkinter import filedialog
import pandas as pd
import os
def select_file():
    file_path = filedialog.askopenfilename(filetypes=[("Excel files", "*.xlsx")])
    file_entry.delete(0, tk.END)
    file_entry.insert(0, file_path)

def select_output_dir():
    output_dir = filedialog.askdirectory()
    output_entry.delete(0, tk.END)
    output_entry.insert(0, output_dir)

def generate_files():
    file_path = file_entry.get()
    output_dir = output_entry.get()
    
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    df = pd.read_excel(file_path)
    unique_values = df['استان'].unique()
    
    for value in unique_values:
        filtered_df = df[df['استان'] == value]
        output_file_path = os.path.join(output_dir, f'{value}_output_file.xlsx')
        filtered_df.to_excel(output_file_path, index=False)
    
    result_label.config(text="!استان ها آماده ان")

# Create main window
root = tk.Tk()
root.title("تفکیک کننده استان - مهندس صبور")

# File selection frame
file_frame = tk.Frame(root)
file_frame.pack(pady=10)

file_label = tk.Label(file_frame, text="لطفا فایل اکسل را انتخاب کنید")
file_label.grid(row=0, column=0)

file_entry = tk.Entry(file_frame, width=50)
file_entry.grid(row=0, column=1, padx=10)

file_button = tk.Button(file_frame, text="انتخاب کردن", command=select_file)
file_button.grid(row=0, column=2)

# Output directory selection frame
output_frame = tk.Frame(root)
output_frame.pack(pady=10)

output_label = tk.Label(output_frame, text="لطفا پوشه ای که مد نظر دارید ریخته شود را انتخاب کنید")
output_label.grid(row=0, column=0)

output_entry = tk.Entry(output_frame, width=50)
output_entry.grid(row=0, column=1, padx=10)

output_button = tk.Button(output_frame, text="انتخاب کردن", command=select_output_dir)
output_button.grid(row=0, column=2)

# Generate files button
generate_button = tk.Button(root, text="بزن بریم", command=generate_files)
generate_button.pack(pady=10)

# Result label
result_label = tk.Label(root, text="")
result_label.pack()

root.mainloop()
```
